### 1. EDA(Exploratory Data Analysis) : 탐색적 데이터 분석

- EDA 개념 :
    - 수집한 데이터(raw data)를 다양한 각도에서 관찰하고 이해하는 과정.
    - 즉, 데이터 분석 전, 그래프나 통계적인 방법으로 자료를 직관적으로 바라보는 과정
- 과정 : 
    - 데이터를 전체적오로 살피기 (head, tail 부분 확인)
    - 이상치, 결측치 확인, 수치형/범주형 등 확인
    - 데이터의 각 feature값 관찰
    - feature들 간의 관계를 관찰하여 패턴 확인 (상관관계, 시각화 등)

#### How many Servived? (얼마나 생존했는지 시각화를 통해 확인)

```python
f, ax = plt.subplots(1,2, figsize=(18,8) #1행,2열의 그래프 위치 만들기
                     
# 첫번째 그려지는 그래프에대한 설정
data['Servived'].value_counts().plot.pie(explode=[0,0,1], autopct = '%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Servived')
ax[0].set_ylabel('')
# 두번째 그려지는 그래프에대한 설정
sns.countplot('Sruvived',data=data,ax=ax[1])
ax[1].set_title('Servived')
                     
 plt.show() #그려라
                     
```

```python
- [ 1) value_counts()  : 카테고리 값 세기 ]
- 각 값이 나온 횟수를 센다.
- 1-1) Series인 경우 (Series에 적합)
    - 전체데이터.value_counts() # 각 값의 종류별로의 몇번씩 나왔는지 반환
    - Series 값이 정수,문자열, 카테고리 인 경우, value_counts()메서드로 각 값이 나온 횟수를 셀 수 있다.
    - 각 값이 몇번씩 나왔는지 반환
    
- 1-2) Dataframe 인 경우
    - df[몇번째열].value_coutns() # n번째 열에있는 데이터에서, 각 값의 종류별로 몇번씩 나왔는지 개수 셈 
    - value_counts 메서드가 없다.  따라서 전체가 아닌 각 열마다 별도로 적용해야 함.
        

>참고 : https://datascienceschool.net/01%20python/04.04%20%EB%8D%B0%EC%9D%B4%ED%84%B0%ED%94%84%EB%A0%88%EC%9E%84%EC%9D%98%20%EB%8D%B0%EC%9D%B4%ED%84%B0%20%EC%A1%B0%EC%9E%91.html
        , https://pearlluck.tistory.com/609, https://bigdaheta.tistory.com/46
                
- [ 2) count() : 데이터 개수 세기 ]
    - 데이터프레임.count() : NaN값은 세지 않는다. 각 열로 유효한 원소 개수 반환
```
